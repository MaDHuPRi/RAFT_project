```
import gradio as gr
import os
import cv2
import time

def detect_and_play(image_file):
    # Run YOLOv5 detection and redirect output to output.txt
    !python detect.py --weights /content/yolov5/runs/train/exp/weights/best.pt --source $image_file.name > output.txt 2>&1

    # Add a time delay of 10 seconds
    time.sleep(10)

    # Get the content of the output file
    with open('output.txt', 'r') as file:
        content = file.read()

    # Check if the crash is detected
    if 'crash' in content:
        # If crash is detected, play the crash-detected audio
        audio_path = "/content/yolov5/crash.mp3"
        detection_status = "Crash is Detected"
    else:
        # If crash is not detected, play the other audio
        audio_path = "/content/yolov5/nocrash.mp3"
        detection_status = "Crash is Not Detected"

    # Get the list of folders in /content/yolov5/runs/detect/
    detect_folders = os.listdir("/content/yolov5/runs/detect")

    # Check if any folder is generated
    if not detect_folders:
        return "No detection folder found", None, None

    # Get the last folder by sorting in descending order
    last_folder = sorted(detect_folders, reverse=True)[0]
    # Get the path to the last detection folder
    last_folder_path = os.path.join("/content/yolov5/runs/detect", last_folder)

    # Get all files in the last folder
    files_in_folder = os.listdir(last_folder_path)
    # Find the first image file in the folder
    detected_image_path = None
    for file_name in files_in_folder:
        if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            detected_image_path = os.path.join(last_folder_path, file_name)
            break

    if detected_image_path is None:
        return "No image found in the last folder", None, None

    # Read the detected image
    detected_image = cv2.imread(detected_image_path)

    return audio_path, detected_image, detection_status

# Define the input component (file upload)
input_component = gr.File(label="Upload an image")

# Define the output components (audio player, image display, and text display)
output_audio = gr.Audio(label="Audio")
output_image = gr.Image(label="Detected Image")
output_text = gr.Textbox(label="Detection Status")

# Create the Gradio interface
gr.Interface(fn=detect_and_play, inputs=input_component, outputs=[output_audio, output_image, output_text]).launch()
```
